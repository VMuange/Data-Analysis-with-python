## pivot() and pivot_table() Methods

- The pivot() method reshapes data based on column values.
  Does not support data aggregation.

- pivot_table() method creates a spreadsheet style pivot table.
  Supports data aggregation.

### 1. Pivot() method

```python
import pandas as pd
```

```python
# Read and show dataset
df_gdp = pd.read_excel(r'D:\Python\Data Science\Data Sets\Python for excel users\gdp.xlsx')
df_gdp
```

```python
# Reshape data with .pivot

df_gdp.pivot(index = 'year', columns='country', values = 'gdppc')
```

### 2. pivot_tables()method

```python
# Read and show dataset

df_sales = pd.read_excel(r'D:\Python\Data Science\Data Sets\Python for excel users\supermarket_sales.xlsx')
df_sales
```

```python
# How many females and males spend their money in this supermarket?

# Make a pivot table and add an aggregate function

df_sales.pivot_table(index = 'Gender', aggfunc = 'sum')
```

```python
# Produce the above pivot table but only with quantity and totals columns

df_sales.pivot_table(index = 'Gender',
                     values = ['Quantity', 'Total'],
                     aggfunc = 'sum')
```

```python
# Create a pivot table showing houch much male and female spend on each product line

df_sales.pivot_table(index = 'Gender',
                     columns = 'Product line',
                     values = 'Total',
                     aggfunc = 'sum').round(2)
```

# Data Visualization with Pandas

```python
# Read and show dataset
df_population_raw = pd.read_excel(r'D:\Python\Data Science\Data Sets\Python for excel users\population_total.xlsx')
df_population_raw
```

# 1. Making a pivot table
.pivot() returns reshaped DataFrame organized by index, column, values without aggregation.
### Cleaning the DataFrame

```python
# Dropping null values using the .dropna() function

df_population_raw.dropna(inplace=True)
```

```python
# Create a pivot table using .pivot() to reshape the data.

df_pivot = df_population_raw.pivot(index= 'year', columns='country', values='population')
df_pivot
```

```python
# Select some countries
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
```

```python
df_pivot
```

# Visualisations.
### 1. Line plot

##### Note that we import matplotlib which we use for saving and exporting the pivot table and plots we will make.

```python
import matplotlib.pyplot as plt
```

```python
df_pivot.plot(kind='line', xlabel='Year',
              ylabel = 'Population', 
              title = 'Population (1955-2020)',
             figsize = (8,4))
# Save plot
plt.savefig('my_test.png')

#Show plot
plt.show()
```

### 2. Bar plot

#### 2.1 Single bar plot

```python
# Select 1 year

df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]

# Show new DataFrame

df_pivot_2020
```

##### When we make a bar plot, we have to put text data i the index.

```python
# Transpose DataFrame from rows to columns and viceversa.

df_pivot_2020 = df_pivot_2020.T  # df_pivot_2020.transpose() works the same.
```

```python
df_pivot_2020
```

```python
df_pivot_2020.plot(kind='bar', xlabel = 'Country', 
                   ylabel = 'Population',
                   title = 'Population(2020)', 
                   color = 'orange')

# Save plot
plt.savefig('Population in 2020.png')

#Show plot
plt.show()
```

#### 2.2 Bar plots grouped by 'n' variables.

```python
df_pivot #Show pivot table
```
